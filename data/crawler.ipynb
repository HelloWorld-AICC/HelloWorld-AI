## 근로복지공단

```python
import json
import time
from tqdm import tqdm
import os
import random
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

url = 'https://www.comwel.or.kr/comwel/help/index.jsp?page=1&start_dt=&end_dt=&req_tp=&type=title&key=%EC%99%B8%EA%B5%AD%EC%9D%B8'
file_path = 'raw_data/Web/comwel.json'
data = []

# Selenium 설정
options = webdriver.ChromeOptions()
# options.add_argument('--headless')
# options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--disable-infobars")

try:
    driver = webdriver.Chrome(options=options)
    driver.get(url)

    for i in tqdm(range(1, 67)):
        driver.get(f"https://www.comwel.or.kr/comwel/help/index.jsp?page={i}&start_dt=&end_dt=&req_tp=&type=title&key=%EC%99%B8%EA%B5%AD%EC%9D%B8")
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#contents > table > tbody > tr")))
        rows = driver.find_elements(By.CSS_SELECTOR, "#contents > table > tbody > tr")

        for row in rows:
            img_alt = row.find_element(By.CSS_SELECTOR, "td:nth-child(3) > img").get_attribute("alt") if row.find_elements(By.CSS_SELECTOR, "td:nth-child(3) > img") else None
            if img_alt == "비공개":
                continue

            topic = row.find_element(By.CSS_SELECTOR, "td:nth-child(2)").text.strip()
            title_element = row.find_element(By.CSS_SELECTOR, "td.td01 > a")
            title = title_element.text.strip()
            date = row.find_element(By.CSS_SELECTOR, "td:nth-child(5)").text.strip()

            # 링크 클릭으로 이동
            ActionChains(driver).move_to_element(title_element).click().perform()
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#contents")))

            question = ""
            question_elements = driver.find_elements(By.CSS_SELECTOR, "#contents > table.view02 > tbody > tr > td > p")
            for q in question_elements:
                question += q.text.strip() + "\n"

            answer = ""
            answer_element = driver.find_elements(By.CSS_SELECTOR, "#contents > table.viewAn > tbody > tr:nth-child(2) > td.td02 > div")
            if answer_element:
                answer += answer_element[0].text.strip() + "\n"

            new_data = {
                "title": topic + "\n" + question.strip(),
                "date": date,
                "contents": answer.strip(),
                "url": driver.current_url
            }

            data.append(new_data)

            # 뒤로 가기
            driver.back()
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#contents > table > tbody > tr")))

finally:
    driver.quit()
    # 데이터 저장
    os.makedirs(os.path.dirname(file_path), exist_ok=True)
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

print("Crawling completed.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Selenium 드라이버 설정
driver = webdriver.Chrome()  # 사용 중인 ChromeDriver로 변경


# 데이터 저장할 리스트
all_data = []

try:
    for page_num in range(1, 11):
        driver.get(f"https://www.easylaw.go.kr/CSP/UnScOnhunListRlt.laf?onhunqnaAstSeq=&curPage={page_num}&outmax=10&sch=%EC%99%B8%EA%B5%AD%EC%9D%B8&fsort=")  # 시작 URL 입력
        time.sleep(1)

        # 페이지 QA 수집
        qa_elements = driver.find_elements(By.CSS_SELECTOR, "#Ask_answer_list01 > li")
        
        for i in range(2, len(qa_elements) + 1, 2):
            time.sleep(0.5)
            # 더보기 버튼 클릭
            try:
                more_button = driver.find_element(By.CSS_SELECTOR, f"#Ask_answer_list01 > li:nth-child({i}) > dl > a")
                href = more_button.get_attribute("href")
                driver.execute_script("window.open(arguments[0], '_blank');", href)
                driver.switch_to.window(driver.window_handles[-1])  # 새 창으로 전환
                
                # 세부 QA 데이터 수집
                topic = driver.find_element(By.CSS_SELECTOR, "#search_result > div.vote_list > ul > li.title").text.strip()
                question = driver.find_element(By.CSS_SELECTOR, "#search_result > div.vote_list > ul > li.qa > div > div.ttl").text.strip()
                answer = driver.find_element(By.CSS_SELECTOR, "#search_result > div.vote_list > ul > li.qa > div > div.ans").text.strip()
                
                # 데이터 저장
                all_data.append({
                    "topic": topic.split("\n")[0],
                    "question": question,
                    "answer": answer
                })
                
                # 원래 페이지로 돌아가기
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
            except Exception as e:
                print(f"Error on QA {i}: {e}")
                continue
        
finally:
    # 데이터 저장
    with open('raw_data/Web/easylaw.json', 'w', encoding='utf-8') as f:
        json.dump(all_data, f, ensure_ascii=False, indent=4)

    driver.quit()
```

## 네이버 지식인

```python
## 네이버 지식인
import requests
from bs4 import BeautifulSoup
import json
from tqdm import tqdm
import random

# URL 및 저장 파일 경로 설정
base_url = 'https://kin.naver.com/userinfo/answerList.naver?u=LozJsi3cGuRd5Y5PklLl8GwBQI1V%2BpaEnpueNikIQss%3D&isSearch=true&query=%EC%99%B8%EA%B5%AD%EC%9D%B8&sd=answer&y=0&section=qna&isWorry=false&x=0&page='
file_path = 'raw_data/Web/jisikin.json'
data = []

# 헤더 설정 (봇 탐지를 피하기 위해)
user_agents = [
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "AppleWebKit/537.36 (KHTML, like Gecko)",
    "Chrome/91.0.4472.124 Safari/537.36",
]


try:
    for page in tqdm(range(1, 14)):
        # 페이지 요청
        url = base_url + str(page)
        headers = {"User-Agent": random.choice(user_agents)}
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # HTTP 오류 발생 시 예외

        # 페이지 파싱
        soup = BeautifulSoup(response.text, 'html.parser')

        # 게시글 목록 가져오기
        board_list = soup.select("#au_board_list > tr > td.title > a")
        for index, element in enumerate(board_list):
            try:
                # URL 생성
                href = element['href']
                if not href.startswith("http"):  # 중복된 http 방지
                    detail_url = "https://kin.naver.com" + href
                else:
                    detail_url = href

                # URL 확인
                print("Fetching URL:", detail_url)

                # 상세 페이지 요청 및 파싱
                detail_response = requests.get(detail_url, headers=headers)
                detail_response.raise_for_status()
                detail_soup = BeautifulSoup(detail_response.text, 'html.parser')
                div = detail_soup.select_one('div.endTitleSection')

                # "blind" 클래스 제거 후 텍스트 가져오기
                for span in div.find_all('span', class_='blind'):
                    span.decompose()  # blind 클래스를 포함한 태그 제거

                visible_text = div.get_text(strip=True)

                # 데이터 추출
                title = visible_text
                contents = detail_soup.select_one("#content > div.endContentLeft._endContentLeft > div.contentArea._contentWrap > div.questionDetail")
                answers = detail_soup.select_one("#answer_1 > div.answerDetail._endContents._endContentsText")
                date = detail_soup.select_one("#answer_1 > div.answerInfo > p")

                if title and contents and answers and date:
                    new_text = {
                        "title": title.strip(),
                        "contents": contents.text.strip(),
                        "answers": answers.text.strip(),
                        "date": date.text.strip()
                    }
                    print(new_text)
                    data.append(new_text)

            except Exception as e:
                print(f"{index + 1}번째 게시글 크롤링 중 오류 발생: {e}")

finally:
    # JSON 데이터 저장
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

    print(f"크롤링 완료. 총 {len(data)}개의 데이터를 저장했습니다.")
```

## 고용노동부

```python
import json
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from tqdm import tqdm
import os
import random
from bs4 import BeautifulSoup
import requests

url = 'https://www.moel.go.kr/minwon/fastcounsel/fastcounselList.do'
file_path = 'raw_data/Web/moel2.json'
data = []

# 헤더 설정 (봇 탐지를 피하기 위해)
user_agents = [
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "AppleWebKit/537.36 (KHTML, like Gecko)",
    "Chrome/91.0.4472.124 Safari/537.36",
]

def get_soup(url):
    headers = {"User-Agent": random.choice(user_agents)}
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    return BeautifulSoup(response.text, 'html.parser')


try:
    # 첫 페이지 로드
    for i in tqdm(range(1, 4962)):
        soup = get_soup(url + f"?pageIndex={str(i)}")
        
        # 데이터 테이블 로드
        rows = soup.select(".board_list tbody tr")
        if not rows:
            break
        
        for row in rows:
            try:
                # 제목, 날짜, 상세 링크
                a_tag = row.select_one("td.txt_left > strong > a")
                date = row.select_one("td:nth-child(3)").get_text(strip=True)
                link = f"https://www.moel.go.kr{a_tag['href']}" if a_tag else None
                
                if not link:
                    continue
                
                # 상세 페이지 로드
                detail_soup = get_soup(link)
                title = detail_soup.select_one("#contents_body .b_info dl:nth-child(1) > dd").get_text(strip=True)
                contents = detail_soup.select_one("#contents_body .b_info dl:nth-child(2) > dd").get_text(strip=True)
                
                new_data = {"title": title, "contents": contents, "date": date, "url": link}
                data.append(new_data)
                # 요청 간격 조정
                time.sleep(random.uniform(0.5, 1))

            except Exception as e:
                print(f"Error processing row: {e}")
                continue
finally:
    # 데이터 저장
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)

print("Crawling completed.")
```
